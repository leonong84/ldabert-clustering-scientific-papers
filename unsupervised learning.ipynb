## **Literature Review**

### **Centroid Based Clustering**



Centroid-based clustering groups the data into non-hierarchical clusters. The most popular centroid-based clustering algorithm is k-means. Although efficient, centroid-based algorithms are sensitive to beginning conditions and outliers.

K-Means and other related algorithms locate cluster centres (centroids) and assign data points to the nearest centroid using distance metrics such as Euclidean distance. This method works well when the clusters are compact and have similar densities. K-Means, on the other hand, may have difficulty with non-spherical clusters, fluctuating cluster sizes, and outliers.

**Best suited for:**  Situations in which the clusters are substantially spherical and equally proportioned. It is most effective when the data can be partitioned into clusters with roughly similar variances.

### **Spectral Clustering**



The connectedness between the data points is used to create the clustering in spectral clustering, a variation of the clustering technique. It forecasts the data into lesser dimensions of space to cluster the data points using the eigenvalues and eigenvectors of the data matrix. It is built on the concept of a graph representation of data, where the nodes represent the data points, and the edges show how similar the data points are to one another.

**Best Suited For**: Spectral clustering is well-suited for tasks where the data has a graph-like structure or when clusters are not necessarily spherical and may have complex relationships.

### **Distribution-based Clustering**


The distributed-based clustering technique separates the data into Gaussian-like distributions with the chance of a point belonging to a distribution decreasing as one gets away from the distribution's centre. The model clusters the data by fitting it to the likelihood of each point belonging to the same distribution. The distribution might be either normal or Gaussian. When there are a limited number of distributions and incoming data is fitted into them to maximise data distribution, the Gaussian distribution becomes increasingly important. (Google, 2022). Distributed clustering operates on two distinct architectural frameworks which are homogeneous datasets and heterogeneous dataset. They refer to datasets where each local place possesses identical attributes and each local location possesses distinct attribute but the linkage between these clusters relies on a shared key attribute, respectively. (Gulati et al. 2015). This clustering technique was evaluated using both external and internal validation indices. External validation indices assess the closeness between the clustering algorithm's output and the appropriate partitioning of the dataset, whereas internal validation indices assess the quality of a clustering structure in the absence of external input. The Jaccard Index, Adjusted Rand Index, Silhouette Index, and Dunn Index were utilised as validation indices in Rodriguez's 2019 study. The outcomes were assessed using a variety of measures, including p-value, mean, standard deviation, maximum value, and average gain in performance. (Rodriguez et al., 2019). This technique is evaluated on various unbalanced datasets and outperforms several other comparable clustering algorithms in several performance indices while clustering data that is skewed. (Wang et al., 2019). Furthermore, although this model performs well on synthetic data and clusters of varying sizes, it may encounter issues if the constraints are not employed to restrict the model's complexity (Priy, 2023). Another limitation of this technique is that the complexity and significance of applications can increase when dealing with data sets that encompass many types of information, such as text, numeric values, and non-numeric value. (Kargupta et al., 2001) Distribution-based clustering algorithms are predominantly developed for numerical data and the cluster is dependent on statistical feature, whereby the concept of distance or similarity can be objectively assessed (Januzaj et al, 2004). So, it may not be suitable in NLP task such as topic modelling due to several reasons. First, the representation of NLP data into text involves the use of feature spaces with high dimensionality. In a given text, every word or phrase can be considered as an unique dimension, leading to a dataset characterised by significant sparsity. Dimensionality reduction techniques like as Principal Component Analysis (PCA) are often employed in distribution-based approaches (Kargupta et al., 2001). However, it is important to note that PCA may not fully capture the fundamental linguistic features. Second, Non-numeric data refers to information that is not expressed in numerical form but rather consists of symbols and text. Distribution-based clustering methods primarily utilise numerical features and distance measures, which are not readily applicable to textual data. The process of transforming textual data into numerical representations using techniques such as TF-IDF or word embeddings is feasible; nevertheless, it is important to note that these methods may not fully capture the underlying semantic meanings of words. According to the study conducted by (Qiaser et al. 2018), the TF-IDF method lacks the ability to assess the semantic meaning of text inside documents, limiting its utility to the lexical level. Furthermore, it lacks the capability to verify word co-occurrences. Third, Distribution-based methods assume that data follows specific probability distributions (Kargupta et al., 2001). The assumptions behind NLP data may not always be applicable, as the distribution of words or subjects in text typically exhibits significant irregularity (Indu et al., 2016). Distribution-based clustering approaches are particularly suitable for numerical data that conforms to specific statistical distributions. For example, distribution-based clustering is a commonly employed technique in the fields of image processing and computer vision, particularly for the purpose of image segmentation. The study conducted by (Lu et al. 2021) introduces a framework for distributed clustering algorithms designed to effectively manage a large volume of supervoxels. The authors also showcase an image segmentation system developed using this methodology. In the field of finance, it is common for financial data, such as stock prices and market returns, to have statistical distributions. The research conducted by (Kargupta et al. 2001) demonstrates the effectiveness of the Distribution-based algorithm in analysing time series data, including daily stock data. ### **Grid Based Clustering**


Grid-based clustering algorithms quantize object regions into a finite number of cells and build a grid structure on which all clustering operations are done using a multi-resolution grid data structure. STING, WaveCluster, and CLIQUE are examples of grid-based clustering approaches utilised in high-dimensional data space. The statistical parameters may be employed in a top-down, grid-based technique to compute the confidence interval indicating the relevance of the cell to a particular query. (Ginni, 2021). Several assessment methods, including the Silhouette coefficient, Dunn index, Davies-Bouldin index, and Calinski-Harabasz index were employed in Guo's paper published in the journal in 2018 to evaluate the clustering algorithm. The Silhouette coefficient evaluates how well each data point fits into its assigned cluster based on distance to neighbouring cluster points as well as distance to points inside its own cluster. The Dunn index measures cluster compactness and separation by dividing the shortest intercluster distance by the greatest intracluster distance. The Davies-Bouldin index computes the average similarity between each cluster and its most comparable cluster based on the distance between their centroids. The Calinski-Harabasz index computes the ratio of within-cluster to between-cluster dispersion, with a higher value suggesting more clustering. Grid-based clustering is extensively used in many industries because it returns clusters of any shape, is robust to noise and does not require prior knowledge of the number of clusters. However, because of its high temporal complexity, it does not work well with large datasets and requires calculating densities for most data points one by one, restricting its performance. (Guo et al., 2021). Besides, grid based algorithm necessitates the determination of an optimal size for the grid structure. In a situation that the grid size is very big, it is possible for two or more clusters to merge into a singular entity. In instances when the grid size is significantly reduced, it is possible for a cluster to be subdivided into many sub-clusters. Hence, the determination of an appropriate grid size is a significant challenge in grid clustering methodologies (Edla et al., 2012).

In the context of topic modelling, it is common for the data to be in the form of textual data. Conversely, Grid-based clustering methods are generally intended for the processing of numerical or geographical data (Wang et al., 1997). Therefore, it may not possess the necessary capabilities to effectively process textual information. Topic modelling necessitates the comprehension of the semantic significance of words, phrases, or sentences. In contrast, grid-based clustering is a methodology that centres on the division and arrangement of data according to their closeness or similarity, as discussed by (Hireche et al. 2020). Grid-based clustering algorithms do not explicitly take into account semantic information, which results in their limited effectiveness in capturing the inherent meaning of textual material.



The tasks related to topic modelling frequently encounter challenges associated with high-dimensional data, which are characterised by issues such as the high dimensional feature space and feature sparsity concerns (Onan et al., 2016).  For example, each dimension signifies a distinct linguistic a feature or attribute. Grid-based clustering algorithms often encounter the curse of dimensionality, a phenomenon in which the quality of clustering diminishes as the number of dimensions grows. Grid-based clustering is not well-suited for topic modelling tasks that require text data with a high number of dimensions. According to (Sia et al., 2020), the use of grid-based or other non-centroid-based hierarchical methods led to inferior performance outcomes and necessitated the adjustment of a greater number of hyperparameters in task related to textual data.


Another limitation seen in grid-based clustering algorithms is the absence of diagonal boundaries. This algorithm include the projection of data onto a grid, which is divided into a predetermined number of cells based on particular parameters. These cells represent the resulting clusters which means they can only capture clusters with horizontal or vertical boundaries (Hireche et al., 2020). In Natural Language Processing (NLP) tasks, clusters may exhibit complex and non-linear boundaries that cannot be effectively captured by grid-based clustering algorithms. This limitation may result in the loss of important cluster structures in text data.


Grid-based clustering algorithms are particularly effective in domains where the data can be represented as a grid or where the spatial relationships between data points are important. For example, Several clustering methods (Agrawai et al., 1998; Leung et al., 2005; Ng et al., 1994) employ the grid-based algorithm to compress extensive source data and mitigate computational complexity. The authors of the study conducted by (Wei et al., 2007) utilised grid-based clustering technique consisting of two phases in order to segment network traffic data. In addition, STING, which is a grid-based clustering algorithm that partitions the dataset hierarchically. has been shown that is appropriate for handling data types such as uncertain and time series data type. Uncertain data refers to data that contains noise, which introduces the possibility of deviation from the intended, exact or initial values. On the other hand, time series data is a form of data that consists of a sequence of data points, which are organised and indexed according to a certain time order or time intervals. ### **K-Means with LDA-Bert hybrid**


Latent Dirichlet Allocation (LDA) is one of the most used topic modeling techniques that can automatically detect topics from a vast collection of text documents (George & Sumathy, 2023). However, LDA-based topic models alone do not always yield promising results. Clustering, an effective unsupervised machine learning algorithm, is extensively used in applications like extracting information from unstructured textual data and topic modeling. A hybrid model combining Bidirectional Encoder Representations from Transformers (BERT) and LDA in topic modeling with clustering based on dimensionality reduction has been studied in detail (George & Sumathy, 2023). The complexity of clustering algorithms increases with the number of features, prompting the use of PCA, t-SNE, and UMAP-based dimensionality reduction methods.

In the realm of manufacturing, the combination of K-means and LDA has been employed to facilitate the automated annotation and classification of manufacturing webpages, thereby enhancing the intelligence of supplier discovery and knowledge acquisition tools (Yazdizadeh Shotorbani et al., 2016). Furthermore, in the biomedical domain, where text data plays a pivotal role, topic modeling techniques have been developed to address the redundancy in biomedical text documents and to discover precise topics. Techniques combining inverse document frequency and machine learning fuzzy k-means clustering algorithms have been proposed to enhance the quality of text mining in this domain (Rashid et al., 2019).

```python
# Parser
parser = en_core_sci_lg.load()
parser.max_length = 7000000 #Limit the size of the parser, how much text the parser can handle at once, prevent overwhelming

def spacy_tokenizer(sentence):
    ''' Function to preprocess text of scientific papers
        (e.g Removing Stopword and puntuations)'''
    mytokens = parser(sentence)
    mytokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens ] # transform to lowercase and then split the sentence
    mytokens = [ word for word in mytokens if word not in stopwords and word not in punctuations ] #remove stopsword and punctuation
    mytokens = " ".join([i for i in mytokens])
    return mytokens
```

## **Modelling**

The workflow for our project depends on the goals and the specific challenges of the dataset at hand. Our approach is:

image.png

1. Dimensionality Reduction: Often, datasets have high-dimensional features that can be noisy or correlated. Dimensionality reduction techniques like PCA, t-SNE, or UMAP can help reduce the dimensionality of the data, which can improve the performance of clustering algorithms by focusing on the most important features and reducing the "curse of dimensionality." Additionally, some clustering algorithms, such as KMeans, work better in lower-dimensional spaces due to their reliance on distances.

2. Clustering: Once the dimensionality of the data is reduced, clustering algorithms are applied to the reduced data to identify patterns or groups.

3. Visualization: The resulting clusters are visualized to represent the data in 2 dimensions. Techniques like t-SNE or UMAP are often used for this purpose because they can capture non-linear structures and are optimized for visual representation. Other than that, topwords for each cluster are also identified to provide insights.

In other cases, clustering might be performed directly on the original high-dimensional data if it's believed that all features are important and there's a risk of losing critical information during dimensionality reduction. After clustering, dimensionality reduction can then be applied just for visualization purposes.

In summary:
- If the primary goal is to get the best clustering performance, then dimensionality reduction is often done before clustering.
- If the primary goal is visualization, then clustering might be done first on the original data, followed by dimensionality reduction to achieve a 2D or 3D representation for visualization.

**Vectorization step**

We will vectorize the abstract column using TF-IDF since we are interested in clustering based on the presence and importance of specific terms. The few advatanges TF-IDF provided are:


1.   **Term Weighting**: TF-IDF not only considers the frequency of a term in a given document (Term Frequency, TF) but also how unique or rare the term is across all documents (Inverse Document Frequency, IDF). This weighting scheme can highlight words that are more important or distinctive in a document.
2.   **High Interpretability**: The vectors produced by TF-IDF have a clear interpretation. Each dimension corresponds to a specific word in the vocabulary, and the value indicates the importance of that word in the document relative to the entire corpus. This is not always the case with methods like word embeddings, where the semantics of individual dimensions are not clear.
3.   **Simple and Efficient**: TF-IDF is straightforward to understand and implement. It's computationally efficient and doesn't require the iterative training processes that some other methods (like neural embeddings) require.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.85)

# Fit and transform the abstracts to produce the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['abstract'].fillna(''))

# Display the shape of the resulting matrix and a sample of the features (words)
tfidf_shape = tfidf_matrix.shape
tfidf_features = tfidf_vectorizer.get_feature_names_out()[:10]

tfidf_shape, tfidf_features
```

1.   The explained variance increases sharply initially, indicating that the first few components capture a significant amount of the total variance in the data.
2.   As the number of components increases, the growth in cumulative explained variance starts to slow down, indicating diminishing returns from adding more components.
3.   There isn't an "elbow" point in the plot, hence it is difficult to pinpoint exact how many components to retain.

Therefore, we will just keep 95% of the variance.

### **Grid Based Clustering**

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import umap.umap_ as umap
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML asgm/processed-data.csv')
abstracts = df['abstract']

# Random select 1000 samples
abstracts = abstracts.sample(n=1000, random_state=42)
```

```python
# vectorizer
grid_tfidf_vectorizer = TfidfVectorizer(max_df=0.85)

# Fit and transform the abstracts to produce the TF-IDF matrix
grid_tfidf_matrix = grid_tfidf_vectorizer.fit_transform(abstracts.fillna('')).toarray()

#grid_tfidf_shape = grid_tfidf_matrix.shape
```

```python
# Dimension reduction using UMAP
grid_reducer = umap.UMAP()
grid_umap_embeddings = grid_reducer.fit_transform(grid_tfidf_matrix)
```

```python
results = []

# Define a range of values for num_grids_per_dimension and similarity_threshold
num_grids_values = [2, 3, 4, 5, 6, 7, 8, 9, 10] #increase -> the grid getting finer, more cluster, may oversegmentation; decrease -> less cluster, mayoversimplyfy data
similarity_threshold_values = [0.5, 0.6, 0.7, 0.8] #i-> clustering strict, more cluster ; d->permissive, less cluster

for num_grids_per_dimension in num_grids_values:
    for similarity_threshold in similarity_threshold_values:


        #num_grids_per_dimension = params['num_grids_per_dimension']
        grid_size = (grid_umap_embeddings.max(axis=0) - grid_umap_embeddings.min(axis=0)) / num_grids_per_dimension

        # Create a dictionary to store points in each grid cell
        grid_cells = defaultdict(list)
        for i, x in enumerate(grid_umap_embeddings):
          grid_coords = tuple((x / grid_size).astype(int))
          grid_cells[grid_coords].append(i)


        # Compute the mean vector for each grid cell
        grid_means = {grid: np.mean(grid_tfidf_matrix[indices], axis=0) for grid, indices in grid_cells.items()}

        # Create a similarity matrix between grid cells based on cosine similarity
        grid_mean_vectors = list(grid_means.values())
        similarity_matrix = cosine_similarity(grid_mean_vectors)

        # Apply a threshold to determine which cells are connected
        #similarity_threshold = param['similarity_threshold']
        connected_cells = similarity_matrix > similarity_threshold

        # Perform clustering based on connected grid cells
        clusters = defaultdict(list)
        visited = set()
        cluster_id = 1
        for i in range(len(grid_mean_vectors)):
            if i in visited:
                continue
            queue = [i]
            while queue:
                current = queue.pop(0)
                visited.add(current)
                clusters[cluster_id].extend(grid_cells[tuple(grid_cells.keys())[current]])
                neighbors = np.where(connected_cells[current])[0]
                for neighbor in neighbors:
                    if neighbor not in visited:
                        queue.append(neighbor)
                        visited.add(neighbor)
            cluster_id += 1


        labels = np.zeros(len(grid_umap_embeddings))


        for cluster_id, points in clusters.items():
            labels[points] = cluster_id

        # Calculate silhouette score
        grid_sil_score = silhouette_score(grid_umap_embeddings, labels)

        # Append the results to the list
        result = {
            'num_grids_per_dimension': num_grids_per_dimension,
            'similarity_threshold': similarity_threshold,
            'silhouette_score': grid_sil_score
        }
        results.append(result)

# Print the results
for result in results:
    print(f"num_grids_per_dimension: {result['num_grids_per_dimension']}, similarity_threshold: {result['similarity_threshold']}, Silhouette Score: {result['silhouette_score']}")

# Find the best result
best_result = max(results, key=lambda x: x['silhouette_score'])
print("\nBest Parameters:")
print(f"num_grids_per_dimension: {best_result['num_grids_per_dimension']}, similarity_threshold: {best_result['similarity_threshold']}")
print(f"Best Silhouette Score: {best_result['silhouette_score']}")
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = 6  # Adjust based on elbow method

lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda_topic_distributions = lda.fit_transform(tfidf_matrix)
```


The number of topics from LDA and the number of clusters in K-Means do not necessarily have to be the same, but they can be related.

- LDA Topics: The number of topics in LDA represents underlying themes or subjects in the dataset. Each document can be represented as a mixture of these topics.

- K-Means Clusters: The number of clusters in K-Means represents distinct groups of data points based on their similarity.

When LDA topic distributions are used as input to K-Means, we are essentially clustering documents based on their topic similarities. In this context:

- If number of K-Means clusters is same as LDA topics, the clusters might correspond closely to individual topics.

- If choosing a different number of K-Means clusters, documents are clustered based on combinations of topics or more granular subtopics.

Considerations:

- Same Number: If the number of clusters matches the number of topics, each cluster might (ideally) represent documents predominantly about one topic. However, this isn't guaranteed, as K-Means doesn't know about the "meaning" of the topics.

- More Clusters: If there are more clusters than topics, there might be more nuanced groupings of documents. For example, if one LDA topic represents "technology," separate K-Means clusters might further distinguish between "software technology" and "hardware technology."

- Fewer Clusters: If there are fewer clusters than topics, multiple topics are being grouped together. Using the above example, "software technology" and "hardware technology" might be grouped together into a broader "technology" cluster.

Therefore, which approach to use depends on the specific goals of our analysis:

- To group documents by broad themes, choose the same number or fewer clusters than LDA topics.

- To group documents into more specific groupings, choose more clusters than LDA topics.

These words can be considered as "domain-specific stop words." Just as common stop words (e.g., "and," "the," "is") are often removed from text because they don't provide specific meaning, domain-specific stop words can also be removed if they don't contribute significantly to distinguishing between topics or clusters in our specific dataset.

```python
# Loading dataset
df = pd.read_csv('/content/drive/MyDrive/ML asgm/processed-data.csv')
df = df.sample(n=1000, random_state=42)

# Remove the domain-specific stop words
domain_stopwords = ["propose", "algorithm", "method", "model", "use", "study", "paper",
                    "result", "analysis","present", "approach", "discuss", "introduce",
                    "provide", "evaluate", "application","perform", "technique",
                    "experiment", "conclusion", "objective", "review", "literature",
                    "contribute", "describe", "research", "find", "show", "work",
                    "system", "datum", "data","base", "based", "performance", "set",
                    "problem", "feature", "solution"]

# Remove domain-specific stopwords from the abstracts
df['abstract'] = df['abstract'].apply(lambda x: ' '.join([word for word in x.split()
                                      if word.lower() not in domain_stopwords]))

# Initialize a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.85)

# Fit and transform the abstracts to produce the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['abstract'].fillna(''))

tfidf_features = tfidf_vectorizer.get_feature_names_out()[:10]

tfidf_shape, tfidf_features
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = 6  # Adjust based on elbow method

lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda_topic_distributions = lda.fit_transform(tfidf_matrix)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
abstracts = df['abstract'].tolist()
bert_embeddings = model.encode(abstracts, show_progress_bar=True)
```

```python
#Combine the LDA topic distributions and BERT embeddings
#for each document.  This will give a new representation
#that includes information  from both LDA topics and BERT embeddings.

combined_features = np.hstack((lda_topic_distributions, bert_embeddings))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Use the elbow method to find the optimal number of clusters
distortions = []
K = range(1, 20)
for k in K:
    kmeanModel = KMeans(n_clusters=k, n_init="auto")
    kmeanModel.fit(combined_features)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

UMAP, which stands for Uniform Manifold Approximation and Projection, is a dimensionality reduction technique that is particularly well-suited for visualizing high-dimensional datasets. It is similar in purpose to techniques like t-SNE and PCA, but offers several advantages:

Flexibility: UMAP can be used for both visualization (reducing dimensions to 2D or 3D) and for general non-linear dimension reduction (reducing to any number of dimensions).

Performance: UMAP is computationally efficient, often faster than t-SNE, especially on larger datasets.

Consistency: While t-SNE can sometimes provide different visualizations on different runs if the random seed changes, UMAP tends to be more consistent.

Preservation of Global Structure: Unlike t-SNE, which mainly focuses on preserving local structure, UMAP balances preserving both the local and global structures in the data.

Generalization: UMAP can be used not just for dimensionality reduction, but also for other tasks, such as clustering, anomaly detection, and more.

Using UMAP instead of PCA:
We can use UMAP instead of PCA for dimensionality reduction. While PCA is linear and focuses on capturing the maximum variance in the data, UMAP is non-linear and can capture more complex patterns and relationships.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import umap.umap_ as umap

# Dimension reduction using UMAP
reducer = umap.UMAP()
umap_embeddings = reducer.fit_transform(combined_features)

# Clustering with 6 clusters on UMAP embeddings
kmeans = KMeans(n_clusters=6)
clusters = kmeans.fit_predict(umap_embeddings)

# Calculate silhouette score
sil_score = silhouette_score(umap_embeddings, clusters)
print(f"Silhouette Score: {sil_score}")

# Visualization
plt.figure(figsize=(6, 3))
plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=clusters, cmap='viridis', s=50)
plt.colorbar()
plt.title(f'UMAP Projection of 6 Clusters (Silhouette Score: {sil_score:.3f})')
plt.show()
```

```python
# Number of clusters and topics (1 topic per cluster to get the top words)
num_clusters = len(set(clusters))
num_topics = 1

top_words_per_cluster = {}

# Initialize LDA model
lda = LatentDirichletAllocation(n_components=num_topics, max_iter=10, random_state=42)

for cluster_id in range(num_clusters):
    # Subset the data for the cluster
    cluster_data = tfidf_matrix[clusters == cluster_id]

    # Apply LDA
    lda.fit(cluster_data)

    # Extract top words for the dominant topic
    topic_words = {}
    for topic_idx, topic in enumerate(lda.components_):
        top_words = [tfidf_vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:][::-1]]
        topic_words[f"Cluster {cluster_id} Topic {topic_idx}"] = top_words

    top_words_per_cluster[cluster_id] = top_words

# Convert to DataFrame
df_cluster_top_words = pd.DataFrame.from_dict(top_words_per_cluster, orient='index', columns=[f'Term {i}' for i in range(1, 11)])
df_cluster_top_words.reset_index(inplace=True)
df_cluster_top_words.rename(columns={"index": "Cluster ID"}, inplace=True)

# Adjust the Cluster ID to start from 1
df_cluster_top_words['Cluster ID'] = df_cluster_top_words['Cluster ID'] + 1

# Remove the index
df_cluster_top_words.set_index('Cluster ID', inplace=True)
df_cluster_top_words
```

After removing the domain-specific stopwords, the keywords in the clusters are becoming more obvious, and we can assign appropriate topics to the clusters.

```python
# Mapping of cluster IDs to labels
cluster_labels = {
    1: "Human-Robot Interaction",
    2: "Mathematical Modeling",
    3: "Energy Systems",
    4: "Cloud & Mobile Networking",
    5: "Medical Imaging & Treatment",
    6: "Multimedia Processing"
}

df_cluster_top_words.reset_index(inplace=True)
df_cluster_top_words.rename(columns={"index": "Cluster ID"}, inplace=True)

# Rename the Cluster ID to the label using the mapping
df_cluster_top_words['Cluster Label'] = df_cluster_top_words['Cluster ID'].map(cluster_labels)

# Set the 'Cluster Label' as index
df_cluster_top_words.set_index('Cluster Label', inplace=True)

# Drop the Cluster ID column
df_cluster_top_words.drop(columns=['Cluster ID'], inplace=True)
df_cluster_top_words
```

The silhouette score is a clustering analysis statistic that measures how near each data point in one cluster is to data points in neighbouring clusters. Its value varies from -1 to 1. The following are the Silhoutte Score Interpretations: 1: The clusters are distinct and well separated from one another. 0: This indicates that the clusters are overlapping. -1: Data points were allocated to the incorrect clusters. The elbow method on the other hand is also another approach to obtain the optimal number of cluster on the clustering algorithm. This method considers the ideal number of clusters to be the point at which the within-cluster sum of squares (WCSS) begins to level out and create an elbow shape.

## **Conclusion**

K-Means clustering with LDA-BERT has the best result on clustering the scientic papers based on abstract from the rest of the algorithms. From our work, we had observed that the vectorization, dimensionality reduction methods and clustering algorithms used has impact on the results.

All of the modeling methods employed in our study incorporated TF-IDF vectorization, with the notable exception of the LDA-BERT hybrid model. In this hybrid approach, LDA is harnessed for topic modeling, while BERT is leveraged for feature enrichment and extraction by embedding words to provide a more profound contextual understanding. This amalgamation of LDA and BERT allows the LDA-BERT hybrid model to exploit the interpretability and topic discovery capabilities of LDA, while simultaneously benefiting from the contextual richness offered by BERT embeddings. This synergistic approach not only facilitates the identification of latent topics within the text data but also enhances the representation of documents with more informative and context-aware features. However, the LDA-BERT method is prone to a phenomenon where certain domain-specific stopwords appear extensively throughout the cluster. Addressing this challenge requires an in-depth domain analysis to identify and subsequently remove these domain-specific stopwords.

UMAP has also been shown to be a better dimensionality reduction technique when dealing with text data. This phenomenon can be attributed to several key factors. Firstly, UMAP is a nonlinear dimensionality reduction technique, while PCA is a linear method. Text data often exhibits complex, nonlinear relationships between words and documents due to the intricate semantic and contextual nuances present in language. UMAP's ability to capture these nonlinear patterns makes it better suited for preserving the inherent structure and relationships within textual data. Secondly, UMAP is designed to preserve both local and global structures in the data, which is particularly valuable for text clustering. In text data, words or phrases can have local associations with specific topics or themes, as well as global relationships that encompass broader contextual meanings. UMAP's capacity to conserve both fine-grained local information and overarching global structures allows it to more effectively represent the intricate interplay of words and topics in text.

In summary, our study has revealed that K-Means clustering with LDA-BERT, in conjunction with UMAP dimensionality reduction, has delivered superior results when applied to clustering scientific papers based on their abstracts. This combination of techniques has exhibited remarkable performance in unveiling meaningful clusters and patterns within the text data. However, it's important to note that the LDA-BERT method is susceptible to a phenomenon where certain domain-specific stopwords proliferate throughout the clusters. Addressing this challenge necessitates a comprehensive domain study to identify and subsequently remove these domain-specific stopwords. These findings emphasize the importance of considering domain-specific nuances in text analysis tasks, as they can significantly impact the quality of clustering outcomes and the overall success of NLP projects.
